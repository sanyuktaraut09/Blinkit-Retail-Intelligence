```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned/blinkit_cleaned.csv")

df.head()
```

```python
total_sales = df["Sales"].sum()
print(f"Total Sales: {total_sales:.2f}")
```

```python
average_sales = df["Sales"].mean()
print(f"Average Sales: {average_sales:.2f}")
```

```python
sales_by_item = df.groupby("Item Type")["Sales"].sum().sort_values(ascending=False)

sales_by_item
```

```python
sales_by_item.plot(kind="bar", figsize=(12,6))

plt.title("Sales by Item Type")
plt.xlabel("Item Type")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

### Business Insight 1

- Fruits & Vegetables generated the highest sales among all product categories.
- Snack Foods is the second-highest revenue contributor.
- Seafood has the lowest sales, indicating lower customer demand.
- Blinkit should prioritize inventory and promotional campaigns for high-performing categories while reviewing the assortment strategy for low-performing categories.

```python
sales_by_outlet = df.groupby("Outlet Type")["Sales"].sum().sort_values(ascending=False)

sales_by_outlet.plot(kind="bar", figsize=(8,5))

plt.title("Sales by Outlet Type")
plt.xlabel("Outlet Type")
plt.ylabel("Total Sales")
plt.xticks(rotation=20)

plt.show()
```

```python
sales_by_size = df.groupby("Outlet Size")["Sales"].sum().sort_values(ascending=False)

sales_by_size.plot(kind="bar", figsize=(7,5))

plt.title("Sales by Outlet Size")
plt.xlabel("Outlet Size")
plt.ylabel("Total Sales")

plt.show()
```

```python
sales_by_location = df.groupby("Outlet Location Type")["Sales"].sum().sort_values(ascending=False)

sales_by_location.plot(kind="bar", figsize=(7,5))

plt.title("Sales by Location Tier")
plt.xlabel("Location")
plt.ylabel("Total Sales")

plt.show()
```

```python
fat_sales = df.groupby("Item Fat Content")["Sales"].mean()

fat_sales.plot(kind="bar", figsize=(6,5))

plt.title("Average Sales by Fat Content")
plt.ylabel("Average Sales")

plt.show()
```

```python
year_sales = df.groupby("Outlet Establishment Year")["Sales"].sum()

year_sales.plot(kind="line", marker="o", figsize=(10,5))

plt.title("Sales by Outlet Establishment Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")

plt.show()
```

# Feature Engineering

```python
import numpy as np

np.random.seed(42)

df["Profit Margin (%)"] = np.random.randint(8, 31, size=len(df))
```

```python
df["Profit"] = df["Sales"] * df["Profit Margin (%)"] / 100
```

```python
df["Discount (%)"] = np.random.choice(
    [0, 5, 10, 15, 20],
    size=len(df),
    p=[0.20, 0.25, 0.30, 0.15, 0.10]
)
```

```python
df["Inventory Status"] = np.random.choice(
    ["In Stock", "Low Stock", "Out of Stock"],
    size=len(df),
    p=[0.75, 0.20, 0.05]
)
```

```python
df["Sales Category"] = pd.cut(
    df["Sales"],
    bins=[0, 75, 150, 300],
    labels=["Low", "Medium", "High"]
)
```
